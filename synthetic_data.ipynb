```python
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

from main import simulate_profile

%matplotlib inline
```

```python
num_profiles = 10000
init_age = 16
age_max = 96

X = []
for _ in range(num_profiles):

    # Make synth screening history.
    X.append(simulate_profile(init_age, age_max))
    
X = np.array(X)
X.shape
```

```python
v, c = np.unique(X[X != 0], return_counts=True)
c, v
```

```python
c / sum(c)
```

```python
np.random.seed(0)
idx = np.random.choice(range(num_profiles), size=20, replace=False)

_, axes = plt.subplots(nrows=len(idx) // 3, ncols=3, figsize=(15, 15))
for num, axis in enumerate(axes.ravel()):

    axis.plot(X[idx[num]], '-o')

    axis.set_yticks(range(0, 5))
    axis.set_yticklabels(['Death', 'N0', 'L1', 'L2', 'C4'], fontsize=18)

    axis.set_xticks(np.linspace(0, 80, 6, dtype=int))
    axis.set_xticklabels(np.linspace(16, 96, 6, dtype=int), fontsize=18)

    axis.set_ylabel('State', fontsize=20)
    axis.set_xlabel('Years', fontsize=20)

plt.tight_layout()
```
